```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
a=print(func4([1,2,3,4],[2,4,6,8]))
```

```python
import time
start_time=time.time()
func4([1,2,3,4],[2,4,6,8])
end_time=time.time()
execution_time=end_time-start_time
print("Execution time:", execution_time, "seconds")
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
import time
start_time=time.time()
func5(16)
end_time=time.time()
execution_time=end_time-start_time
print("Execution time:", execution_time, "seconds")
```

```python
# Exercise 6
def func8(A):
    return func8_helper(A, 0, len(A) - 1)

def func8_helper(A, i, j):
    n = j - i + 1
    if n == 2 and A[i] > A[j]:
        A[i], A[j] = A[j], A[i]
    if n > 2:
        m = n // 3
        func8_helper(A, i, j - m)
        func8_helper(A, i + m, j)
        func8_helper(A, i, j - m)

    return A

```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum = 0
  for i in arr:
    sum += i
  return sum
  print(sum)

```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr)
```

```python
import time
start_time = time.time()
sum_arr(arr)
end_time = time.time()
execution_time = end_time - start_time
print("Execution time:", execution_time, "seconds")
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  max = arr[0]
  min = arr[0]
  for i in arr:
    if i > max:
      max = i
    elif i < min:
      min = i
    return max, min
    print(max, min)
    print(max_min)
    print(arr)
    print(max_min(arr))
```
